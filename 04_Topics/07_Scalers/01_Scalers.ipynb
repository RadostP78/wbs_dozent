# Scaler

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python

matplotlib.style.use('fivethirtyeight')
np.random.normal
```

# Create a dataset

```python
df = pd.DataFrame({
    "x1" : np.concatenate([ np.random.normal(20, 2, 1000 ) ,  np.random.normal(1 , 2 , 25 )  ])  ,     # mean: 20 , std: 2, 1000 points
    
    "x2" : np.concatenate([ np.random.normal(30, 2, 1000 ) ,  np.random.normal(50 , 2 , 25 )  ])       # mean: 30 , std: 2, 1000 points
})
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

# Build the scalers

```python
# MinMax Scaler 
scaler = preprocessing.MinMaxScaler()
minmax_df = scaler.fit_transform(df)
minmax_df = pd.DataFrame(minmax_df, columns = ["x1", "x2"])

# StandardScaler
scaler = preprocessing.StandardScaler()
standard_df = scaler.fit_transform(df)
standard_df = pd.DataFrame(standard_df, columns = ["x1", "x2"])


# Robustscaler
scaler = preprocessing.RobustScaler()
robust_df = scaler.fit_transform(df)
robust_df = pd.DataFrame(robust_df, columns = ["x1", "x2"])

```

# Plot and Compare

```python
fig, (ax1, ax2, ax3, ax4)  = plt.subplots(ncols= 4, figsize = (20,5))

ax1.set_title("Before Scaling")
sns.kdeplot(df["x1"], ax = ax1, color = "r")
sns.kdeplot(df["x2"], ax = ax1, color = "b")


ax2.set_title("Min_Max Scaler")
sns.kdeplot(minmax_df["x1"], ax = ax2, color = "r")
sns.kdeplot(minmax_df["x2"], ax = ax2, color = "b")


ax3.set_title("Standard Scaler")
sns.kdeplot(standard_df["x1"], ax = ax3, color = "r")
sns.kdeplot(standard_df["x2"], ax = ax3, color = "b")


ax4.set_title("Robust Scaler")
sns.kdeplot(robust_df["x1"], ax = ax4, color = "r")
sns.kdeplot(robust_df["x2"], ax = ax4, color = "b")


```
